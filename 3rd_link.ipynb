```python
                                                                      # 2nd project




import random

def guess_the_number():
    # Step 1: Computer picks a random number between 1 and 100
    number_to_guess = random.randint(1, 100)

    # Step 2: Initialize attempt counter
    attempts = 0
    guessed_correctly = False

    print("Welcome to Guess the Number Game!")
    print("I have selected a number between 1 and 100. Can you guess it?")

    # Step 3: Game loop
    while not guessed_correctly:
        # Step 4: Take input from the user
        try:
            user_guess = int(input("Enter your guess: "))
        except ValueError:
            print("Please enter a valid number!")
            continue

        attempts += 1

        # Step 5: Provide feedback on the guess
        if user_guess < number_to_guess:
            print("Too low! Try again.")
        elif user_guess > number_to_guess:
            print("Too high! Try again.")
        else:
            guessed_correctly = True
            print(f"Congratulations! You've guessed the number in {attempts} attempts.")

# Run the game
guess_the_number()
```

```python
                                                            #3rd project
import random

# Guess the Number Game logic
def guess_the_number():
    number_to_guess = random.randint(1, 100)
    attempts = 0
    hint_given = False
    max_attempts = 5  # Number of attempts before hint is given

    print("Welcome to Guess the Number!")
    print("Guess a number between 1 and 100:")

    while True:
        try:
            user_guess = int(input("Enter your guess: "))
        except ValueError:
            print("Please enter a valid number!")
            continue

        attempts += 1

        # Give hints after max attempts
        if attempts >= max_attempts and not hint_given:
            if number_to_guess % 2 == 0:
                print("Hint: The number is even!")
            else:
                print("Hint: The number is odd!")
            hint_given = True

        if user_guess < number_to_guess:
            print("Too low! Try again.")
        elif user_guess > number_to_guess:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You guessed the number in {attempts} attempts.")
            break  # End the game after correct guess

# Run the game
guess_the_number()





```

```python
                                                         #




# Function to play the game where computer guesses the user's number
def computer_guess_number():
    print("Welcome to the Guess the Number game!")
    print("Think of a number between 1 and 100, and I will try to guess it.")

    low = 1
    high = 100
    attempts = 0
    user_feedback = ''

    # Keep guessing until the computer guesses the correct number
    while user_feedback != 'c':
        guess = (low + high) // 2  # Guess the middle number
        attempts += 1
        print(f"Is your number {guess}?")
        user_feedback = input("Enter 'h' if my guess is too high, 'l' if it's too low, and 'c' if I guessed correctly: ").lower()

        if user_feedback == 'h':
            high = guess - 1  # Adjust the range to a lower number
        elif user_feedback == 'l':
            low = guess + 1  # Adjust the range to a higher number
        elif user_feedback != 'c':
            print("Invalid input! Please enter 'h', 'l', or 'c'.")

    print(f"Yay! I guessed your number {guess} in {attempts} attempts.")

# Run the game
computer_guess_number()




```

```python
import random

# Options & Emojis
options = ["rock", "paper", "scissors"]
emoji = {"rock": "🪨", "paper": "📄", "scissors": "✂️"}

# Scores
user_score = 0
computer_score = 0
rounds_played = 0

print("🎮 Welcome to Rock, Paper, Scissors Game!")
print("Type 'exit' anytime to quit.\n")

while True:
    user = input("Choose rock, paper, or scissors: ").lower()

    if user == "exit":
        print("\n👋 Game Over!")
        print(f"📊 Total Rounds: {rounds_played}")
        print(f"🧑 You: {user_score} | 🤖 Computer: {computer_score}")
        break

    if user not in options:
        print("⚠️ Invalid choice. Try again!\n")
        continue

    computer = random.choice(options)
    print(f"🤖 Computer chose: {computer} {emoji[computer]}")

    rounds_played += 1

    if user == computer:
        print("😐 It's a tie!\n")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("🎉 You won this round!\n")
        user_score += 1
    else:
        print("💔 You lost this round!\n")
        computer_score += 1
```

```python
import random

# Word list
words = ['python', 'program', 'hangman', 'challenge', 'developer']

# Choose a random word
word = random.choice(words)
guessed = ['_'] * len(word)
used_letters = []
attempts = 6  # lives

print("🎮 Welcome to Hangman!")
print("Guess the word: " + ' '.join(guessed))

while attempts > 0 and '_' in guessed:
    guess = input("Guess a letter: ").lower()

    if not guess.isalpha() or len(guess) != 1:
        print("⚠️ Please enter a single letter!\n")
        continue

    if guess in used_letters:
        print("🔁 You already guessed that letter!\n")
        continue

    used_letters.append(guess)

    if guess in word:
        print("✅ Correct!\n")
        for idx, letter in enumerate(word):
            if letter == guess:
                guessed[idx] = guess
    else:
        print("❌ Wrong! You lost a life.\n")
        attempts -= 1

    print("Word: " + ' '.join(guessed))
    print(f"Lives left: {attempts}")
    print("Used letters:", ', '.join(used_letters), "\n")

# Game over result
if '_' not in guessed:
    print("🎉 Congratulations! You guessed the word:", word)
else:
    print("💔 You lost. The word was:", word)
```

```python
import time

def countdown_timer(seconds):
    while seconds > 0:
        mins = seconds // 60
        secs = seconds % 60
        timer = f"{mins:02d}:{secs:02d}"
        print(timer, end="\r")  # \r rewrites the line
        time.sleep(1)
        seconds -= 1

    print("⏰ Time's up! ⏰")

# User input
user_time = int(input("Enter time in seconds: "))
countdown_timer(user_time)
```

```python
import random
import string

def generate_password(length):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

# User input
length = int(input("Enter password length: "))
print("🔐 Your generated password is:", generate_password(length))
```
